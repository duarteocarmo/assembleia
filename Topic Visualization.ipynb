```python
import gensim
import pickle
import pyLDAvis.gensim

dictionary = gensim.corpora.Dictionary.load('nlp_models/dictionnary.gensim')
corpus = pickle.load(open('nlp_models/corpus.pkl', 'rb'))
lda = gensim.models.ldamodel.LdaModel.load('nlp_models/model5.gensim')
lda_display = pyLDAvis.gensim.prepare(lda, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_display)
```
